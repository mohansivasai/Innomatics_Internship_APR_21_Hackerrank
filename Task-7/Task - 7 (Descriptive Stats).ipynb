<h2>Mean (Arithmetic Mean): </h2>
<blockquote> 
    <b><i>
     Mean is average of group of numbers, in other words, mean is the sum of numbers in the group divided by count of total numbers in the group.
        </i> </b>
   </blockquote>


$$ 
\bar{X} = \frac{1}{N} \sum_{i=1}^{N} x_{i} = \frac{x_{1}+x_{2}+....+x_{i}}{N}
$$


<blockquote>
    <b>Where,</b>
    
$ \bar{X} = Mean \\
       N  = Total \ number \ of \ observations \ in \ a \ given \ set\\
    x_{i} = dataValues
$
<h3>Properties of Mean:</h3>
    <ul>
        <li> The Mean Value is impacted by the Outliers.</li>
        <li> The Order of Complexity for Mean is <b>O(n)</li>
    </ul>
</blockquote>    

<h6> Finding the Mean of the column "Mthly_HH_Income"</h6>

```python
def user_mean(x):
    sum = 0
    n = len(x)
    for i in x:
        sum+= i
    mean = sum/n
    return mean
```

<h2>Median:</h2>
<blockquote>
    <b>Median is the middle number in the sorted list of numbers either in ascending order or descending order.</b>
 </blockquote>

$$
Median = \left\{
    \begin{array}\\
        (\frac{n+1}{2})^{th} observation & \mbox{if } \ n \ is \ odd\\
        \frac{\frac{n}{2}^{th} observation + (\frac{n}{2} +1)^{th} observation} {2} & \mbox{if } \ n \ is \ even \\ 
    \end{array}
\right.
$$

<blockquote>
    <b>Where,</b>
    
$ N  = Total \ number \ of \ observations \ in \ a \ given \ set\\ $
<h3>Properties of Median:</h3>
    <ul>
        <li> <i>The Median is robust to Outliers.</i></li>
        <li> The Order of Complexity for Median is <b>O(nlogn)</b></li>
    </ul>

</blockquote>  

<h6> Finding the Median of the column "Mthly_HH_Expense"</h6>

```python
def user_median(x):
    sort_arr = sorted(x)
    n = len(x)
    if n%2 != 0:
        median = sort_arr[n//2]
    else:
        median = (sort_arr[n//2]+sort_arr[n//2-1])/2
    return median    
```

         Types of modes:<br>

<h5><b> The Mode value is Robust to Outliers.</b></h5>
    
 </blockquote>



<h6> Finding the Mode of the column "Mthly_HH_Income"</h6>

```python
def user_mode(arr):
    d = {}
    for n in arr:
        if n in d:
            d[n] += 1
        else:
            d[n] = 1
    val = list(d.values())
    mode = [key for key,value in d.items() if value==max(val)]
    return mode[0]
```

<h2> Variance:</h2>
<blockquote>
    <b>Variance is a measure of how data points differ from the mean i.e It is a measure of how far a set of data (numbers) are spread out from their mean (average) value.</b>
</blockquote>

$$
 \sigma^{2} = \frac{1}{N}\sum_{i=1}^{N} (x_{i}-{\mu})^{2}
$$


<blockquote>
    <b>Where,</b>
    
$ \sigma^{2} = \text{Variance} \\
          \mu = \text{Mean of the given data} \\
         N  = \text{Total number of observations in a given set}\\
       x_{i} = \text{data Values}
$   

<h6> Finding the Variance of the column "Mthly_HH_Income"</h6>

```python
def user_var(arr): 
    n = len(arr)
    mean = sum(arr)/n
    s = 0
    for i in range(n):
       s += (arr[i]-mean)**2 
    var = s/(n-1) ## for sample we often use (n-1) in denominator
    return var
```

```python
print(user_var(data['Mthly_HH_Expense']))
```

# Standard Deviation :
<blockquote>
    <b>In statistics, the standard deviation is a measure of the amount of variation or dispersion of a set of values. A low standard deviation indicates that the values tend to be close to the mean of the set, while a high standard deviation indicates that the values are spread out over a wider range</b>
    </blockquote>


$$
   \sigma = \sqrt{\sigma^{2}} = \sqrt{variance}
$$

<blockquote>
    <b>Where,</b>
    
$ 
\sigma = \text{Standard Deviation} \\
\sigma^{2} = \text{Variance} \\
$   

##### Finding the Standard Deviation of the column "Mthly_HH_Income"

```python
def user_std(x):
    arr = x
    n = len(arr)
    mean = sum(arr)/n
    s = 0
    for i in range(n):
        s += (arr[i]-mean)**2
    var = s/(n-1) ## we often use (n-1) in case of taken sample.
    std = var**0.5
    return std
```

```python
print(user_std(data['Mthly_HH_Income']))
```

# Correlation:
<blockquote>
    <b> Correlation shows the strength of a relationship between two variables and is expressed numerically by the correlation coefficient. The correlation coefficient's values range between -1.0 and 1.0.<br>

A perfect positive correlation means that the correlation coefficient is exactly 1. A perfect negative correlation means that two assets move in opposite directions, while a zero correlation implies no linear relationship at all.</b>

$$ 
 \rho = \frac{covariance(x,y)}{\sigma_{x}*{\sigma_y}}
$$
<b> where,</b>
$$
  \text{Covarriance:} \ \  Cov(x,y) = \frac{1}{n} \sum_{i=1}^{N} (x_{i}-\mu_{x})(y_{i}-\mu_{u})  
$$

<blockquote>
    <h5> here,</h5>
    $
      \sigma_{x} = \text{Standard Deviation of 'X'} \\
      \sigma_{y} = \text{Standard Deviation of 'Y'} \\
      \mu_{x} = \text{Mean of 'X'} \\
       \mu_{y} = \text{Mean of 'Y'} \\
    $
    


<h6> Finding the Correlation coefficient between "Mthly_HH_Income "and "Mthly_HH_Expense"</h6>

```python
def user_corr(x,y):
    n = len(x)
    ux = sum(x)/n
    uy = sum(y)/n
    std_x = (sum([(i-ux)**2 for i in x])/n)**0.5
    std_y = (sum([(j-uy)**2 for j in y])/n)**0.5

    iterable = map(lambda x,y:(x,y),x,y)
    s = 0
    for i,j in iterable:
        s += (i-ux)*(j-uy) 
    covar = s/n
    corr_coef = covar/(std_x*std_y)
    return corr_coef
```

```python
income = data['Mthly_HH_Income'].values
expense = data['Mthly_HH_Expense'].values
print(user_corr(income,expense))
```

# Normal Distribution:

<blockquote>
    <ul>
        <li><b>The normal distribution is a continuous probability distribution that is symmetrical on both sides of the mean.</b></li>
        <li><b>It is also called as " Gaussian Distribution" and occurs most frequently in nature.</b></li>
        <li><b>The Areab under the Curve is equal to '1'.
        <li>For a perfectly normal distribution the mean, median and mode will be the same value, visually represented by the peak of the curve</li>
    </ul>
            
<h6>Mathematical representaion</h6>
            
$$
 X \sim N(\mu,\sigma^{2})
 $$
            <h5>where</h5> 
 $ X = \text{Random variables} \\
  \sim = \ \text{follows} \\
  \mu = \text{Mean}\\
  \sigma^{2} = \text{Standard Deviation}
 $
<h6>For a Stanadard Normal distribution , the mean is zero and standard deviation is '1'</h6>
            
            
$$ X \sim N(0,1)$$
            
           
 <h6> The Probability Density Function of a Normal Distribution is given by </h6>
   
$$ 
P(X=x) = \frac{1}{\sqrt{2\pi}\sigma}*e^ \left\{ \frac{-(x-\mu)^{}2}{2\sigma^{2}}  \right\}
$$

# Features of Normal Distribution:

### 1. It is Symmentric:
<ul>
<li>A normal distribution comes with a perfectly symmentrical shape. This means that the distribution curve have exactly half of the values to the left of center and exactly half the values to the right.</li>
<li>
   The Symmetrical Shape occurs when one-half of the obseravtions falls onn each side of the curve</li> 
    <li>The Curve looks like <b>"Bell-Shaped"</b></li>

<h3>2. The Mean, Median, and Mode are equal:</h3>
<ul>
<li>The middle point of a normal distribution is the point with the maximum frequency, which means that it possesses the most observations of the variable. <li>The measures of central tendecies are usually equal in a perfectly (normal) distribution.
</ul>

### 3. Empirical rule:

In normally distributed data, there is a constant proportion of distance lying under the curve between the mean and specific number of standard deviations from the mean.<br> For example:<ul> <li><b>68.25% </b>of all cases fall within +/- <b>one </b>standard deviation from the mean. <li><b>95%</b> of all cases fall within +/- <b>two</b> standard deviations from the mean <li> <b>99%</b> of all cases fall within +/- <b>three</b> standard deviations from the mean.



### 4. Skewness and kurtosis:

Skewness and kurtosis are coefficients that measure how different a distribution is from a normal distribution.<br> <h6>Skewness:</h6>It measures the symmetry of a normal distribution.<br> <h6>Kurtosis:</h6>It measures the thickness of the tail ends relative to the tails of a normal distribution.i.e Kurtosis tells you the height and sharpness of the central peak, relative to that of a standard bell curve.


# Positively Skewed & Negatively Skewed Normal Distribution

### Skewness: 


In statistics, skewness is a measure of the asymmetry of the probability distribution of a random variable about its mean. In other words, skewness tells you the amount and direction of skew (departure from horizontal symmetry). The skewness value can be positive or negative, or even undefined. If skewness is 0, the data are perfectly symmetrical, although it is quite unlikely for real-world data. As a general rule of thumb:
<ul>
<li>If skewness is less than -1 or greater than 1, the distribution is highly skewed.
<li>If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
<li>If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

### Types of Skewness in Normal Distribution:

a) Positively Skewed Normal Distribution <br>
b) Negatively Skewed Normal Distribution

### Positively Skewed Normal Distribution:

<blockquote>
    <ul>
        <li>
            A Positive-skewed distribution has a long right tail. Positive-skewed distributions are also called <b>"Right-skew distributions"</b>. Thats because there are Outliers (long tail) in the positive direction on the number line. The mean is also to the right of the peak.
            <li>

The value of skewness for a positively skewed distribution is greater than zero.
                
                
<h5> Effect of Measures of Central Tendency on Positive skew:</h5>
                
<b> <ul><li> For a Positively Skewed Normal Distribution, the mean is influenced by the presence of outliers at the upper end of the datas range.
    <li> In this case, the mean is highest while Mode is lowest and Median Lies between both.

### Negatively Skewed Normal Distributoin:

<blockquote>
    <ul>
        <li>
            A Negatively-skewed distribution has a long left tail. It is also called <b>"Left-skew distributions"</b>. Thats because there are outliers (long tail) in the negative direction on the number line. The mean is also to the left of the peak.
            <li>

                
                
<h5> Effect of Measures of Central Tendency on Positive skew:</h5>
                
<b> <ul><li> For a Negatively Skewed Normal Distribution, the mean is influenced by the presence of outliers at the lower end of the datas range. 
    <li> In this case, the mean is lowest while Mode is highest and Median Lies between both.

### Effect of Central Tendencies on Skewed (Positive / Negative) Normal  Distributions:

<h5>To summarize, generally if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean. </h5>

# Q-Q Plot (Quantile - Quantile):

<blockquote>
<ul>
<li><b> A Q-Q Plot is a plot of quantiles of two distributions against each other or a plot based on the estimates of quantiles.
    <li>It is a graphical technique for determining if two data sets come from populations with a common distribution, If both sets of quantiles came from the same distribution, we should see the points forming a line that's roughly straight
    <li>The q-q plot is similar to a probability plot. For a probability plot, the quantiles for one of the data samples are replaced with the quantiles of a theoretical distribution.
    <li> It is a Way to test the Normality of the given distribution.

<blockquote>
    <b>If the Values or dots lie on the reference line i.e 45-degree line, then Y is said to be follow a Normal distribution.

### Example of Q-Q plot for Normal distribution: 

```python
plt.figure(figsize=(10,5))
normal_dist = np.random.normal(loc = 20, scale = 5, size=10000)
plt.subplot(1,2,1)
sns.distplot(normal_dist)
plt.subplot(1,2,2)
stats.probplot(normal_dist, dist="norm", plot=plt)
plt.grid()
```

# Box -Cox Transformation:

<blockquote>
    <b> The box-cox normality plot is a plot of the correlation coefficients for various values of the lambda parameter.
    <ul>
        <li> In Machine Learning, we use Q-Q plot and Box-Cox plot to transform given distributions to Normal distribution.
        <li>When you are dealing with real-world data, you are going to deal with features that are heavily skewed. Transformation technique is useful to stabilize variance, make the data more normal distribution-like, improve the validity of measures of association.
        <li>The problem with the Box-Cox Transformation is estimating lambda. This value will depend on the existing data, and should be considered when performing cross validation on out of sample datasets. Make sure to estimate lambda according to the training dataset.

#### Example : Consider the Pareto Distribution:

```python
pareto_rv = np.random.pareto(a = 2, size = 1000)
sns.distplot(pareto_rv)
plt.show()
```

#### Lets check the Normality before applying Box-Cox Transformation:

```python
stats.probplot(pareto_rv, dist="norm", plot=plt)
plt.grid()
```

#### Applying the Box-Cox Transformation:

```python
x_t, l = stats.boxcox(pareto_rv)
print("Lambda:",l)
stats.probplot(x_t, dist="norm", plot=plt)
plt.grid()
```

<b> Summary : We have observed, how Pareto dostribution(pareto_rv) has converted into Normal distribution(x_t) usinig Box-Cox and we have checked the Normality using Q-Q plot.
